What is the most a customer has spent on a single purchase and what is their customerId?

```python
q = ''' 
SELECT CustomerId, MAX(Total) as max
FROM Invoice
GROUP BY CustomerId
ORDER BY max DESC'''
pd.read_sql(q,engine)
```
